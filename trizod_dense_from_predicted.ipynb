```python
import pickle
import pandas as pd
import ast
```

```python
true_dense = pd.read_csv('data/trizod_test_set_dense.tsv', sep='\t').dropna()
true_dense['contact_density'] = true_dense['contact_density'].apply(ast.literal_eval)
```

```python
with open('trizod_test_set_predictions_final_cnn.pkl', 'rb') as f:
    predicted_dense = pickle.load(f)
    
predicted = predicted_dense['test_preds']
keys = predicted_dense['test_ids']
```

```python
true_dense.head()
```

```python
# flatten the predicted and the keys
predicted_flat = []
keys_flat = []
for i in range(len(keys)):
    for j in range(len(keys[i])):
        predicted_flat.append(predicted[i][j])
        keys_flat.append(keys[i][j])
```

```python
predicted_flat[:2]
```

```python
trizod_predicted_dense = []
for i in range(len(keys_flat)):
    key_name = keys_flat[i]
    pred = predicted_flat[i].tolist()
    # get the true density and check how long the 'contact_density' array is
    true_density = true_dense[true_dense['PDBchain'] == key_name]['contact_density'].values[0]
    pred = pred[:len(true_density)]
    trizod_predicted_dense.append({'PDBchain': key_name, 'missing': 0, 'contact_density': pred})
    
# to dataframe
trizod_predicted_dense = pd.DataFrame(trizod_predicted_dense)
trizod_predicted_dense
```

```python
# save the dataframe as tsv
trizod_predicted_dense.to_csv('data/trizod_test_set_dense_predicted.tsv', sep='\t', index=False)
```
